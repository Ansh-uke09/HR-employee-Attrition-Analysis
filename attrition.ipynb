```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Gender', hue='Attrition', data=df)

# Add labels to each bar
for p in ax.patches:
    count = int(p.get_height())
    ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')

# Add titles and labels
plt.title("Attrition by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python

yes_attrition = df[df['Attrition'] == 'Yes']

# Count job roles among attrition cases
job_counts = yes_attrition['JobRole'].value_counts()

# Plot pie chart
plt.figure(figsize=(7,7))
plt.pie(job_counts, labels=job_counts.index, autopct='%1.1f%%', startangle=140 )
plt.title('Attrtion (Yes) by Job Rolei').set_backgroundcolor('lightblue')   
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.
plt.show()
```

```python
yes_attrition = df[df['Attrition'] == 'Yes']

# Count number of attritions by age
age_counts = yes_attrition['JobSatisfaction'].value_counts().sort_index()

# Plot line chart
plt.figure(figsize=(8, 5))
plt.plot(age_counts.index, age_counts.values, marker='o', linestyle='-')
plt.title('Attrition (Yes) by JobSatisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Number of Employees with Attrition = Yes')
plt.grid(True)
plt.show()
```

```python
yes_attrition = df[df['Attrition'] == 'Yes']

# Plot bar chart with Gender on x-axis and EducationField as hue
plt.figure(figsize=(10, 6))
ax=sns.countplot(data=yes_attrition, x='Gender', hue='EducationField')
for p in ax.patches:
    count = int(p.get_height())
    ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')
plt.title('Attrition = Yes: Counts by Gender and Education Field')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Education Field', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
ax =sns.countplot(data=df, x='Attrition', hue='MaritalStatus')
for p in ax.patches:
    count = int(p.get_height())
    ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')
# Add labels and title
plt.title('Attrition by Marital Status')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.legend(title='Marital Status')
plt.show()
```

```python
yes_attrition = df[df['Attrition'] == 'Yes']
sns.histplot(data=df, x='MonthlyIncome', hue='Gender',binrange=(0, 20000), kde=True, stat='count', alpha=0.5)
plt.figure(figsize=(10, 6))
plt.hist(yes_attrition['MonthlyIncome'], bins=30, alpha=0.5, color='blue', label='Attrition = Yes')
plt.hist(df['MonthlyIncome'], bins=30, alpha=0.5, color='orange', label='All Employees')
plt.legend()
plt.title('Monthly Income Distribution')
plt.xlabel('Monthly Income')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

```python
df_yes = df[df['Attrition'] == 'Yes']
df_no = df[df['Attrition'] == 'No']

plt.figure(figsize=(8, 6))
plt.scatter([1]*len(df_yes), df_yes['Age'], color='red', label='Yes', alpha=0.7)
plt.scatter([0]*len(df_no), df_no['Age'], color='green', label='No', alpha=0.7)

# Customize x-axis
plt.xticks([0, 1], ['No', 'Yes'])
plt.xlabel('Attrition')
plt.ylabel('Age')
plt.title('Scatter Plot: Age vs Attrition')
plt.legend()
plt.grid(True)
plt.show()
```

```python
pivot_table = df.pivot_table(index='JobRole', columns='Attrition', values='Age', aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='.1f')

plt.title('Average Age by Job Role and Attrition')
plt.xlabel('Attrition')
plt.ylabel('Job Role')
plt.show()
```
